```python
import geopandas as gpd
from buildingregulariser import regularize_geodataframe
from pathlib import Path
import time
```

```python
test_data_dir = Path("../data/")
original_buildings = gpd.read_file(test_data_dir / "test_data.gpkg")
```

```python
start = time.time()

regularized_buildings = regularize_geodataframe(
    original_buildings,
    simplify_tolerance=2.0,
)

print(f"{len(original_buildings) / (time.time() - start):.0f} polygons/sec")
```

```python
import folium

# Create a Folium map with desired zoom settings and satellite tiles
m = folium.Map(
    location=[-31.882, 115.929],
    zoom_start=20,
    max_zoom=20,
    tiles="https://mt1.google.com/vt/lyrs=s&x={x}&y={y}&z={z}",
    attr="Google",
)

# Add the buildings layer with a different color
original_buildings.to_crs(epsg=4326).explore(
    m=m,
    name="Original Buildings",
    style_kwds={
        "color": "white",
        "weight": 3,
        "fillOpacity": 0.0,
    },
)
```

```python
# Add the regularized_gdf layer with a distinct color
regularized_buildings.to_crs(epsg=4326).explore(
    m=m,
    name="Regularized Buildings",
    style_kwds={"fillColor": "red", "color": "blue", "weight": 3, "fillOpacity": 0.4},
)


folium.LayerControl().add_to(m)
m
```
